```python
from scripts import unargm as u
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
with open("./data/weights.pkl", "rb") as f:
    w = pickle.load(f)
```

```python
M = []; unmax = []
for wmatrix in w:
    d = wmatrix.shape[1]
    M.append(d)
    out = u.unargm(W=wmatrix)
    unmax.append(out.count(1))
    print(f"{d} done!")
```

```python
plt.scatter(M, unmax)
plt.title("2 epoches resnet18, where matrix W is L x d")
plt.xlabel("MRL dimension (d)"); plt.ylabel("Number of unmaxable classes")
plt.show()
```
