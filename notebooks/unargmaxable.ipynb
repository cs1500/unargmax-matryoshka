```python
# resolve paths
import sys
from pathlib import Path

project_root = Path.cwd().parent
sys.path.append(str(project_root))
```

```python
from scripts import unargm as u
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
file = "resnet50_mrl_epochs01"
```

```python
with open(f"../data/{file}.pkl", "rb") as f:
    w = pickle.load(f)
```

```python
M = []; unmax = []
for wmatrix in w:
    d = wmatrix.shape[1]
    M.append(d)
    out = u.unargm(W=wmatrix)
    unmax.append(out.count(1))
    print(f"{d} done!")
```

```python
plt.figure(figsize=(7, 4))
plt.bar(M, unmax, width=12)
#plt.title("2 epoches resnet18, where matrix W is L x d")
plt.xlabel("MRL dimension (d)"); plt.ylabel("Number of unmaxable classes")
plt.grid(True, alpha=0.3)
plt.tight_layout()
#plt.savefig(f"../figures/{file}.pdf", bbox_inches="tight")
plt.savefig(f"../figures/{file}.pdf", format="pdf")
plt.show()
```

```python
# save results
with open(f"../data/{file}_unmax.pkl", "wb") as f:
    pickle.dump(unmax, f)
```
